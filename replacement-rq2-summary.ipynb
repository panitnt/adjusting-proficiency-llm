```python
import pandas as pd
```

```python
df = pd.read_csv('result/ori-replaced-edited.csv')
df
```

```python
ele = df['Unnamed: 0'].unique()
ele
```

```python
len(ele)
```

```python
df_map = df[['Unnamed: 0', 'Class']].drop_duplicates()
df_map
```

```python
start_df = []

for i, row in df_map.iterrows():
    ele_num = row['Unnamed: 0']
    cls = row['Class']
    df_search = df[df['Unnamed: 0']== ele_num]
#     print(df_search)
    df_score = df_search['Replaced Level'].to_list()
    df_score = sorted(df_score)
    unique_element = df_search['Replaced Element'].unique()
    unique_element_sort = sorted(unique_element)
#     print(ele_num)
#     print(unique_element_sort)
    start_df.append([ele_num, cls, unique_element_sort, df_score])
#     print('-------------------------------------------')
```

```python
# Convert the list of lists to a DataFrame
start_df_df = pd.DataFrame(start_df, columns=['i', 'Class', 'unique_element_sort', 'score_list'])

# Save the DataFrame to a CSV file
# start_df_df.to_csv('result/element/all.csv', index=False)
start_df_df.to_csv('result/element/all-withscore.csv', index=False)
```

## Find unique

```python
new_df = start_df_df
new_df
```

```python
# Assuming new_df is already defined and contains the relevant columns
new_df['unique_element_sort_tuple'] = new_df['unique_element_sort'].apply(tuple)
new_df['score_list_tuple'] = new_df['score_list'].apply(tuple)

# Group by 'Class' and the new tuple column
replacement_counts = new_df.groupby(['Class', 'unique_element_sort_tuple', 'score_list_tuple']).size().reset_index(name='Count')
replacement_counts
```

```python
sort_replacement = replacement_counts.sort_values(by='Count', ascending=False)
sort_replacement
```
